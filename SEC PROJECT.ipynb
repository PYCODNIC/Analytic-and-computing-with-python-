# Use numpy to find the min, max, std, variance and correlation of any 3 columns of your dataset

```python
import numpy as np
selected_columns = maindata[['NMHC(GT)', 'PT08.S2(NMHC)', 'PT08.S3(NOx)']]
min = np.min(selected_columns, axis=0)
print(" MINIMUM: ")
print(min)
max = np.max(selected_columns, axis=0)
print("\n MAXIMUM")
print(max)
std = np.std(selected_columns, axis=0)
print("\n STANDARD DEVIATION")
print(std)
variance = np.var(selected_columns, axis=0)
print("\n VARIANCE")
print(variance)
```

### DEALING WITH MISSING VALUES

```python
#FILLING THE MISSING VALUES IN ALL COLUMNS
maindata = maindata.fillna({'Date': maindata['T'].mean()})
maindata = maindata.fillna({'Time': maindata['T'].mean()})
maindata = maindata.fillna({'CO(GT)': maindata['CO(GT)'].mean()})
maindata = maindata.fillna({'PT08.S1(CO)': maindata['PT08.S1(CO)'].mean()})
maindata = maindata.fillna({'NMHC(GT)': maindata['NMHC(GT)'].mean()})
maindata = maindata.fillna({'C6H6(GT)': maindata['C6H6(GT)'].mean()})
maindata = maindata.fillna({'PT08.S2(NMHC)': maindata['PT08.S2(NMHC)'].mean()})
maindata = maindata.fillna({'NOx(GT)': maindata['NOx(GT)'].mean()})
maindata = maindata.fillna({'PT08.S3(NOx)': maindata['PT08.S3(NOx)'].mean()})
maindata = maindata.fillna({'NO2(GT)': maindata['NO2(GT)'].mean()})
maindata = maindata.fillna({'PT08.S4(NO2)': maindata['PT08.S4(NO2)'].mean()})
maindata = maindata.fillna({'PT08.S5(O3)': maindata['PT08.S5(O3)'].mean()})
maindata = maindata.fillna({'T': maindata['T'].mean()})
maindata = maindata.fillna({'RH': maindata['RH'].mean()})
maindata = maindata.fillna({'AH': maindata['AH'].mean()})
maindata = maindata.fillna({'Unnamed: 15': maindata['Unnamed: 15'].mean()})
maindata = maindata.fillna({'Unnamed: 16': maindata['Unnamed: 16'].mean()})
```
